```python
import numpy
import pandas as pd
```

```python
df=pd.read_csv('startup.csv')
df
```

```python
dummies=pd.get_dummies(df['State'])
dummies
# represent the state columns value with dummy variables
```

```python
merged=pd.concat([df,dummies],axis=1)
merged
```

```python
final=merged.drop(['State','New York'],axis=1)
final
# new can be represented by 00 instead of 001
# new york is removed to avoid dummy variable trap
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
X=final.drop(['Profit'],axis=1)
y=final['Profit']
model.fit(X,y)
```

```python
model.coef_
```

```python
model.coef_
```

```python
import pickle
```

```python
with open('pickle_startup_model','wb') as f:
    pickle.dump(model,f)
```

```python
with open('pickle_startup_model','rb')as f:
    md=pickle.load(f)
```

```python
md.predict([[165349.20,136897.80,471784.10,0,0]])#prediction
```

```python
md.score(X,y)*100,'%'
# Accuracy
```

# One Hot Encoding

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
dfle=df
dfle['State']=le.fit_transform(dfle['State'])
# dfle['State']
dfle
```

```python
X=dfle.drop(['Profit'],axis=1)
Y=dfle['Profit']
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(categorical_features=[3])
```

```python
X=ohe.fit_transform(X).toarray()
X
```

```python
X=X[:,1:]
X
# Avoiding dummy variable trap
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
regressor.score(X_train,Y_train)
```
